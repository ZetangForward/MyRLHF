```python
import matplotlib.pyplot as plt

# 数据
api_counts = [400, 300, 200, 100]
avg_lengths = [64310.48, 49141.69, 32543.4875, 16827.855]

# 绘制柱状图
plt.figure(figsize=(10, 6))
plt.bar(api_counts, avg_lengths, color='skyblue', width=40)  # 设置柱子的宽度

# 添加标题和标签
plt.title('Average Length of APIs for Different Query Counts')
plt.xlabel('Number of Queries')
plt.ylabel('Average Length')

# 设置横坐标刻度
plt.xticks(api_counts)  # 只显示四个数字

# 显示数值标签
for i, (api_count, avg_length) in enumerate(zip(api_counts, avg_lengths)):
    plt.text(api_count, avg_length + 1000, f'{avg_length:.2f}', ha='center', va='bottom')

# 显示图形
plt.tight_layout()  # 调整布局以防止标签被截断
plt.show()
```
